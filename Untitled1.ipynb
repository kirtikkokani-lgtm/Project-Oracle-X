```python
def oracle_decision(price_trend, news_score):
    # logic: जर ट्रेंड वर आहे (+1) आणि न्यूज पण चांगली आहे (>0.05)
    if price_trend == "UP" and news_score > 0.05:
        return "STRONG BUY 💎 (जबरदस्त संधी!)"
    
    # logic: जर ट्रेंड खाली आहे (-1) आणि न्यूज पण वाईट आहे (<-0.05)
    elif price_trend == "DOWN" and news_score < -0.05:
        return "STRONG SELL 🔥 (धोका, बाहेर पडा!)"
    
    # logic: जर दोन्ही विरुद्ध असतील
    elif (price_trend == "UP" and news_score < -0.05) or (price_trend == "DOWN" and news_score > 0.05):
        return "HOLD / WAIT ⏳ (गोंधळ आहे, थांबा!)"
    
    else:
        return "NEUTRAL 😐 (मार्केट शांत आहे)"

# टेस्ट करूया
current_trend = "UP" # समजा चार्ट सांगतोय मार्केट वर जाईल
current_news_score = 0.4404 # मघाशी मिळालेला रिलायन्सचा स्कोअर

final_call = oracle_decision(current_trend, current_news_score)
print(f"Project Oracle-X Decision: {final_call}")
```

```python
import yfinance as yf
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# १. मेंदूचे घटक (Functions)
analyzer = SentimentIntensityAnalyzer()

def get_sentiment_score(text_list):
    total_score = 0
    for text in text_list:
        score = analyzer.polarity_scores(text)
        total_score += score['compound']
    return total_score / len(text_list)

def calculate_rsi(data, window=14):
    delta = data.diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()
    rs = gain / loss
    return 100 - (100 / (1 + rs))

# २. डेटा गोळा करणे (Input)
ticker = "RELIANCE.NS"
news_headlines = [
    "Reliance expands its green energy footprint with new deal",
    "Markets remain volatile due to global inflation",
    "Reliance quarterly profits beat analyst expectations"
]

# ३. प्रोसेसिंग (Processing)
df = yf.download(ticker, period="1mo", interval="1d", progress=False)
df['RSI'] = calculate_rsi(df['Close'])

# किमती आणि RSI सुरक्षितपणे काढणे
current_price = df['Close'].iloc[-1]
current_rsi = df['RSI'].iloc[-1]
if isinstance(current_price, pd.Series): current_price = current_price.iloc[0]
if isinstance(current_rsi, pd.Series): current_rsi = current_rsi.iloc[0]

news_score = get_sentiment_score(news_headlines)

# ४. हायब्रीड लॉजिक (The Oracle Strategy)
print(f"\n--- 🔮 PROJECT ORACLE-X: HYBRID DECISION ENGINE ---")
print(f"Stock: {ticker} | Price: {current_price:.2f} | RSI: {current_rsi:.2f}")
print(f"News Sentiment Score: {news_score:.2f}")
print("-" * 50)

if current_rsi < 40 and news_score > 0.1:
    decision = "🚀 LEGENDARY BUY: Technicals and News are both Positive!"
elif current_rsi > 70 and news_score < -0.1:
    decision = "🔥 LEGENDARY SELL: Overbought and Negative News. Exit Now!"
elif news_score > 0.2:
    decision = "📈 BUY ON NEWS: Sentiment is very strong."
elif current_rsi < 30:
    decision = "🛒 TECHNICAL BUY: Stock is Oversold."
else:
    decision = "⏳ HOLD: No clear signal from Oracle-X."

print(f"FINAL DECISION: {decision}")
print("-" * 50)
```

```python
import requests
from bs4 import BeautifulSoup

def get_live_news(stock_name):
    # Google News वरून बातम्या शोधण्यासाठी URL
    url = f"https://www.google.com/search?q={stock_name}+stock+news&tbm=nws"
    
    # ब्राउझरसारखी रिक्वेस्ट पाठवण्यासाठी headers (जेणेकरून ब्लॉक होणार नाही)
    headers = {'User-Agent': 'Mozilla/5.0'}
    
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    headlines = []
    # Google News च्या हेडलाईन्स शोधणे
    for g in soup.find_all('div', class_='BNeaW vvjwJb AP7Wnd'):
        headlines.append(g.get_text())
    
    return headlines[:5] # पहिल्या ५ महत्त्वाच्या बातम्या घेऊया

# टेस्ट करूया
stock = "Reliance Industries"
print(f"--- 📡 Searching Live News for {stock} ---")
live_news = get_live_news(stock)

for i, news in enumerate(live_news, 1):
    print(f"{i}. {news}")
```
